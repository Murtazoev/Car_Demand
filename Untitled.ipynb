```python
import pandas as pd

data = pd.read_csv('Elite Sports Cars in Data.csv')
data
```

```python
data[['Price' , 'Log_Price']]
```

```python
data.loc[data['Brand'] == 'Nissan' , 'Model'] = 'GT-R'
data.loc[data['Brand'] == 'McLaren' , 'Model'] = '720S'
data.loc[data['Brand'] == 'Chevrolet' , 'Model'] = 'Corvette Z06'
data.loc[data['Brand'] == 'Bugatti' , 'Model'] = 'Chiron'
data.loc[data['Brand'] == 'BMW' , 'Model'] = 'M4 Competition'
data.loc[data['Brand'] == 'Porsche' , 'Model'] = '911 Turbo S'
data.loc[data['Brand'] == 'Aston Martin' , 'Model'] = 'DBS'
data.loc[data['Brand'] == 'Lamborghini' , 'Model'] = 'Huracan'
data.loc[data['Brand'] == 'Ferrari' , 'Model'] = '488 GTB'
data.loc[data['Brand'] == 'Ford' , 'Model'] = 'Mustang GT'
```

```python
data
```

```python
y = data['Log_Price']
X = data[[col for col in data.columns if col != 'Log_Price']]
X
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
encoder = LabelEncoder()

X_obj = X.select_dtypes(include = 'object')

for i in X_obj.columns:
    X_obj[i] = encoder.fit_transform(X[i])
```

```python
X_num = X.select_dtypes(include = 'number')
X_final = pd.concat([X_num.reset_index(drop=True), X_obj.reset_index(drop=True)], axis=1)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_final , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
lreg.fit(X_train , y_train)
```

```python
from sklearn.metrics import f1_score, r2_score, accuracy_score
```

```python
y_pred = lreg.predict(X_test)
r2 = r2_score(y_pred , y_test)
r2
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
```

```python
y1 = encoder.fit_transform(data['Market_Demand'])
X1 = data[[col for col in data.columns if col != 'Market_Demand']]

X1_num = X1.select_dtypes(include = 'number')
X1_obj = X1.select_dtypes(include = 'object')

for i in X1_obj.columns:
    X1_obj[i] = encoder.fit_transform(X1[i])

X1_final = pd.concat([X1_num.reset_index(drop=True), X1_obj.reset_index(drop=True)], axis=1)
X1_train , X1_test , y1_train , y1_test = train_test_split(X1_final , y1 , test_size = 0.2 , random_state = 42)
forest = RandomForestClassifier(n_estimators = 100 , random_state = 42)
forest.fit(X1_train , y1_train)
y_pred_forest = forest.predict(X1_test)

y_pred_forest1 = forest.predict_proba(X1_test)[:,1]

roc_score = roc_auc_score(y1_test , y_pred_forest1)
accuracy = accuracy_score(y1_test , y_pred_forest)
accuracy
```

```python
precision = precision_score(y1_test, y_pred_forest, average='weighted')
recall = recall_score(y1_test, y_pred_forest, average='weighted')
f1 = f1_score(y1_test, y_pred_forest, average='weighted')
precision
```
